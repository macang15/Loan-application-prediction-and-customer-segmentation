### 명시적으로 중복되는 로그 제거

### 추가로 중복의 의미를 갖는 로그를 제거

- 1초 간격으로 연속으로 기록된 로그 제거
- 각 Event를 실행한 횟수(연속성을 배제)에 대한 정보를 얻기 위해
- 동일한 이벤트 로그가 연속으로 기록되었다는 것은 하나의 동작이 연속적으로 기록된 형태일 수 있음.

```python
events = normalized_1d.event.unique().tolist()
events
```

```python
normalized_1d = normalized_1d.reset_index()
normalized_1d.columns = ['org_index', 'user_id', 'event', 'timestamp', 'mp_os', 'mp_app_version', 'date_cd']
normalized_1d.tail()
```

```python
for i in tqdm(range(len(events))):

    # 로그에서 이벤트가 신청완료인 로그만 추출
    normalized_1d_part_i = normalized_1d[normalized_1d['event']==events[i]]

    a = len(normalized_1d)

    # 시간 순에 따라 정렬
    normalized_1d_part_i = normalized_1d_part_i.sort_values('timestamp')
    # 기종의 경우 현재 관심사가 아니므로 잠시 배제
    normalized_1d_part_i.drop(normalized_1d_part_i.columns[[4,5]], axis=1, inplace=True)
    # 오리지널 인덱스(기록된 로그상의 인덱스), 유저번호(고유), timestamp에 대해 이벤트(event)와 생성된날짜(date_cd)의 개수를 파악해본다.
    normalized_1d_part_i_grouped = normalized_1d_part_i.groupby(['org_index', 'user_id', 'timestamp']).count().reset_index()


    # 그룹화한 테이블에서 timestamp 행만 추출하여 임시로 pandas.DataFrame()형태로 저장
    timestamp_for_concat = pd.DataFrame(normalized_1d_part_i_grouped['timestamp'])
    # shift(1)로 데이터를 한 칸 아래로 이동시킴
    timestamp_for_concat = timestamp_for_concat.shift(1)
    # DataFrame에서 shift(1)을 진행할 경우 가장 첫 번째 관측치(로그)에서 NaN값이 발생하므로 이를 기존 값으로 채운다.
    timestamp_for_concat['timestamp'][0] = timestamp_for_concat['timestamp'][1]
    # timestamp_for_concat의 컬럼명을 정의해준다.
    timestamp_for_concat.columns = ['timestamp1']


    # 앞서 계산을 위해 생성해주었던 timestamp_for_concat를 기존 테이블에 가로로 합쳐준다.
    test_ith_event = pd.concat([normalized_1d_part_i_grouped, timestamp_for_concat], axis=1)
    # 이후에 event와 date_cd의 경우 의미있는 컬럼은 아니므로 우선 timestamp 계산을 위해 drop해준다.
    test_ith_event.drop(test_ith_event.columns[[3,4]], axis=1, inplace=True)


    # 두 컬럼의 차이를 비교하기 이전에 각 컬럼의 값들이 문자열 형태이므로 pandas의 datetime 형식으로 변경해준다.
    test_ith_event['timestamp']=pd.to_datetime(test_ith_event['timestamp'])
    test_ith_event['timestamp1']=pd.to_datetime(test_ith_event['timestamp1'])


    test_ith_event['diff']=test_ith_event['timestamp']-test_ith_event['timestamp1']
    test_ith_event['diff_sec'] = test_ith_event['diff'].apply(lambda x: x.total_seconds())


    test_ith_event['diff_sec'].iloc[0] = 3


    # 'res_diff'라는 새로운 컬럼에 앞 관측치(직전의 로그)와의 시간 차가 2초 미만인 경우에 True를 저장한다.(나머지는 False 저장)
    test_ith_event['res_diff'] = test_ith_event['diff_sec'].apply(lambda x: 0 <= x < 2)


    # temp 테이블에서 user_id 행만 추출하여 임시로 pandas.DataFrame()형태로 저장
    user_id_for_concat = pd.DataFrame(test_ith_event['user_id'])
    # shift(1)로 데이터를 한 칸 아래로 이동시킴
    user_id_for_concat = user_id_for_concat.shift(1)
    # DataFrame에서 shift(1)을 진행할 경우 가장 첫 번째 관측치(로그)에서 NaN값이 발생하므로 이를 -1로 대체하면 이후 0이상의 값을 갖는 어떤 user_id에서 빼더라도 항상 양수값을 갖는다.(이후에 0인 경우 연속하는 로그의 유저 변화 여부를 확인하기 위해 사용 예정)
    user_id_for_concat['user_id'].iloc[0] = -1
    # user_id_for_concat 컬럼명을 정의해준다. user_id1
    user_id_for_concat.columns = ['user_id1']

    # 앞서 계산을 위해 생성해주었던 user_id_for_concat 기존 테이블에 가로로 합쳐준다.
    calc_table = pd.concat([test_ith_event, user_id_for_concat], axis=1)

    # 해당 이벤트 로그에서 전 이벤트 로그의 user_id값의 차를 기록하는 컬럼을 'is_continuous_user'라고 정의하자.
    # 만약 이 컬럼의 값이 0인 경우 해당 로그는 이전 로그와 같은 사용자가 기록한 로그라는 의미를 갖는다.
    calc_table['is_continuous_user']=calc_table['user_id']-calc_table['user_id1']


    # 이제 계산에 활용한 컬럼들을 모두 drop해주려 한다.
    # timestamp1, diff, diff_sec, user_id1이 제거된다.
    calc_table.drop(calc_table.columns[[3,4,5,7]], axis=1, inplace=True)


    # is_continuous_user 부분을 Boolean 형태로 변경해주자.
    calc_table['is_continuous_user'] = calc_table['is_continuous_user'].apply(lambda x : True if x == 0 else False)

    # 제거할 인덱스 찾아내기 (set 형태 멤버연산자 사용)
    tmp = calc_table[calc_table['res_diff'] == True]
    tmp = tmp[tmp['is_continuous_user'] == True]
    index_to_eliminate = set(tmp['org_index'])

    # for check
    b = len(index_to_eliminate)
    # if b != 0:
    # EndLoanApply 이벤트에서 연속으로 같은 유저에게 기록된 로그 삭제하기
    normalized_1d['to_eliminate'] = normalized_1d['org_index'].apply(lambda x: x in index_to_eliminate)
    normalized_1d = normalized_1d[normalized_1d['to_eliminate']==False]
    normalized_1d = normalized_1d.drop(columns=['to_eliminate'], axis=1)

    # for check
    c = len(normalized_1d)
    
    # 예외처리
    if c != a - b:
      print(f"{i}번째 중복제거 중 문제 발생!")

    print()
    print(f'이벤트 {events[i]}의 중복 제거 전 log_data의 총 로그 개수 : {a}')
    print(f'중복이 의심되는 이벤트 {events[i]}의 로그 개수 : {b}')
    print(f'이벤트 {events[i]}의 중복 제거 후 log_data의 총 로그 개수 : {c}')

normalized_1d.to_csv(f'/content/drive/Shareddrives/빅콘테스트/로그/1D_normalized_log_data({file_version}).csv', index=False)
```

### 버전 변환 함수 정의

3.1.1인 경우 3.01 이런식으로 1) 자릿수 맞추기 (소수점 아래 숫자 두 자리로) 2) . 하나만 포함하도록 너무 minor한 업데이트 삭제

* 앱 버전 + 문자열 이상치도 해결

```python
def normalize_version(data):
  if data == 'nan':
    return 'null'
    #나중에 465 같은 것 len으로 작업해서 안 헷갈리기 위해 null로 바꿈
  if len(data) == 3 or data == '10000':
    return data
  # 버전 업데이트 넘버가 세자리수가 되는 상황은 배제함
  else:
    a,b=data.split('.')[:2]
    if len(b) == 1:
      return (a+'.0'+b)
    else:
      return ".".join(data.split('.')[:2])
```

```python
log_data['mp_app_version']=log_data['mp_app_version'].apply(lambda x:normalize_version(x))
```
