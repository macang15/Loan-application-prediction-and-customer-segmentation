```python
undersample = RandomUnderSampler(sampling_strategy=0.2)
x_under, y_under = undersample.fit_resample(x, y)

X_train = pd.concat([x_under, y_under], axis=1)
print(len(X_train[X_train['is_applied']==1]), len(X_train[X_train['is_applied']==0]))
print(len(y[y.is_applied == 1]), len(y[y.is_applied == 0]))
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=3)
# 4: 0.722
# 6(over 4): 0.727
# 6(over 4) & ver.2 : 0.726
### ver.2 채택

predba_lst = list()
model_lst = [LGBMClassifier(boosting_type='gbdt', class_weight=None, colsample_bytree=1.0,
               importance_type='split', learning_rate=0.5, max_depth=1000,
               min_child_samples=20, min_child_weight=0.001, min_split_gain=0.0,
               n_estimators=1000, n_jobs=-1, num_leaves=126, objective=None,
               random_state=1798, reg_alpha=0.02, reg_lambda=0.2, subsample=1.0,
               subsample_for_bin=200000, subsample_freq=0) for _ in range(10)]


f = 0
for train_index, test_index in kf.split(x_under):
    x_train, x_test = x_under.iloc[train_index], x_under.iloc[test_index]
    y_train, y_test = y_under.iloc[train_index], y_under.iloc[test_index]
    model_lst[f].fit(x_train, y_train)

    y_train_pred = model_lst[f].predict(x_train)
    y_test_pred = model_lst[f].predict(x_test)
    y_whole_pred = model_lst[f].predict(x)
    predba_lst.append(model_lst[f].predict_proba(x)[:,1])

    print(f'========================{f+1}번째 model========================')
    print(f'{f+1}번째 훈련 모델, 훈련 set의 f1_Score:        {f1_score(y_train, y_train_pred)}')
    print(f'{f+1}번째 훈련 모델, 평가 set의 f1_Score:        {f1_score(y_test, y_test_pred)}')
    print(f'{f+1}번째 훈련 모델, 전체 set의 AUC_Score:       {roc_auc_score(y, y_whole_pred)}')
    print(f'{f+1}번째 훈련 모델, 전체 set의 precision_Score: {precision_score(y, y_whole_pred)}')
    print(f'{f+1}번째 훈련 모델, 전체 set의 recall_Score:    {recall_score(y, y_whole_pred)}')
    print(f'{f+1}번째 훈련 모델, 전체 set의 f1_Score:        {f1_score(y, y_whole_pred)}')

    f += 1

f=1
y_pred_df = pd.DataFrame()
for pred in predba_lst:
  y_pred_df[f'pred_{f}'] = pred
  f+=1

y_pred_df['sum'] = 0
col_lst = list(y_pred_df.columns)
col_lst.remove('sum')
for col in col_lst:
  y_pred_df['sum'] += y_pred_df[col] * 0.1

y_pred = y_pred_df['sum'].apply(lambda x: 1 if x >= 0.8 else 0)

print('========================soft voting model: 0.8 over========================')
print(f'전체 set에 대한 voting_AUC_Score:       {roc_auc_score(y, y_pred)}')
print(f'전체 set에 대한 voting_precision_Score: {precision_score(y, y_pred)}')
print(f'전체 set에 대한 voting_recall_Score:    {recall_score(y, y_pred)}')
print(f'전체 set에 대한 voting_f1_Score:        {f1_score(y, y_pred)}')

y_pred = y_pred_df['sum'].apply(lambda x: 1 if x >= 0.7 else 0)

print('========================soft voting model: 0.7 over========================')
print(f'전체 set에 대한 voting_AUC_Score:       {roc_auc_score(y, y_pred)}')
print(f'전체 set에 대한 voting_precision_Score: {precision_score(y, y_pred)}')
print(f'전체 set에 대한 voting_recall_Score:    {recall_score(y, y_pred)}')
print(f'전체 set에 대한 voting_f1_Score:        {f1_score(y, y_pred)}')

y_pred = y_pred_df['sum'].apply(lambda x: 1 if x >= 0.6 else 0)

print('========================soft voting model: 0.6 over========================')
print(f'전체 set에 대한 voting_AUC_Score:       {roc_auc_score(y, y_pred)}')
print(f'전체 set에 대한 voting_precision_Score: {precision_score(y, y_pred)}')
print(f'전체 set에 대한 voting_recall_Score:    {recall_score(y, y_pred)}')
print(f'전체 set에 대한 voting_f1_Score:        {f1_score(y, y_pred)}')

y_pred = y_pred_df['sum'].apply(lambda x: 1 if x >= 0.5 else 0)

print('========================soft voting model: 0.5 over========================')
print(f'전체 set에 대한 voting_AUC_Score:       {roc_auc_score(y, y_pred)}')
print(f'전체 set에 대한 voting_precision_Score: {precision_score(y, y_pred)}')
print(f'전체 set에 대한 voting_recall_Score:    {recall_score(y, y_pred)}')
print(f'전체 set에 대한 voting_f1_Score:        {f1_score(y, y_pred)}')
```
